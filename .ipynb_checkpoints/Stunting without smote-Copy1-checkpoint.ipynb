```python
data = df.iloc[:393,:10]
data = data.drop(columns='Tinggi Badan(cm)')
data
```

```python
# 0 = bukan stunting
# 1 = stunting
data['risiko hamil'] = data['risiko hamil'].apply(lambda value: 0 if value == 1 else 1)
data
```

```python
from sklearn.model_selection import train_test_split

X = data.drop('risiko hamil', axis=1)
y = data['risiko hamil']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

print("Jumlah Data Train : {0}".format(len(X_train)))
print("Jumlah Data Test : {0}".format(len(X_test)))
```

```python
#==================== RF ===================
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(random_state = 42)
classifier.fit(X_train, y_train)
preds = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, recall_score, confusion_matrix

print(f'Accuracy = {accuracy_score(y_test, preds):.2f}\nRecall = {recall_score(y_test, preds):.2f}\n')
cm = confusion_matrix(y_test, preds)
plt.figure(figsize=(8, 6))
plt.title('Confusion Matrix (with SMOTE)', size=16)
sns.heatmap(cm, annot=True, cmap='Blues');
```

```python
#===================== Decision Tree ======================
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
print(f'Accuracy = {accuracy_score(y_test, y_pred):.2f}\nRecall = {recall_score(y_test, y_pred):.2f}\n')
cm = confusion_matrix(y_test, preds)
plt.figure(figsize=(8, 6))
plt.title('Confusion Matrix (with SMOTE)', size=16)
sns.heatmap(cm, annot=True, cmap='Blues');
```
